```python
from sklearn.feature_selection import VarianceThreshold
import pandas as pd

# Simple example data
data = {
'Age':      [22, 45, 31, 67, 18],    # High variance - useful
'Country':  [1,  1,  1,  1,  1],      # Zero variance - useless
'Fare':     [7.25, 71.28, 8.05, 30.0, 10.5] # High variance - useful
}

df = pd.DataFrame(data)
print("Original DataFrame:")    
print(df)
```

```python
# Remove features with variance less than 0.1
selector = VarianceThreshold(threshold=0.1)
df_filtered = selector.fit_transform(df)

print('Original shape:', df.shape)        # (5, 3) - 3 columns
print('After filtering:', df_filtered.shape) # (5, 2) - 'Country' removed!
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Example: We have 4 numeric features
data = {
'Age':    [22, 45, 31, 67, 18],
'Fare':   [7.25, 71.28, 8.05, 30.0, 10.5],
'SibSp':  [1, 0, 3, 0, 1],
'Parch':  [0, 0, 1, 0, 2]
}
df = pd.DataFrame(data)

# Step 1: Scale the data first (PCA needs this!)
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
# Step 2: Apply PCA - reduce from 4 features to 2 components
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

print('Original shape:', df.shape)   # (5, 4) - 4 features
print('After PCA shape:', df_pca.shape) # (5, 2) - 2 components
# See how much information each component keeps
print('Variance explained:', pca.explained_variance_ratio_)
```
